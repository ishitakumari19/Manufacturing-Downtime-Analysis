# Manufacturing Downtime Analysis

This project analyzes manufacturing machine downtime data to identify:
- Failure patterns

## Data Loading

In this section, the dataset is imported and initial records are examined.

```python
import pandas as pd

df = pd.read_csv("manufacturing_downtime_dataset.csv")

df.head()
```

## Data Cleaning

Checking for missing values, duplicate records, and correcting data types.

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# Root Cause Analysis

Root Cause Analysis helps identify the primary contributors to downtime and production losses.

```python
machine_downtime = (
    df.groupby('machine_id')['downtime_minutes']
    .sum()
    .sort_values(ascending=False)
)

print(machine_downtime)
```

## Pareto Analysis (80/20 Rule)

The Pareto principle suggests that a small number of machines often contribute to the majority of downtime.

This analysis identifies the critical machines responsible for most operational disruptions.

```python
cumulative_percentage = (
    machine_downtime.cumsum() / machine_downtime.sum()
) * 100
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10,5))

# Bar chart
machine_downtime.plot(kind='bar', ax=ax1)

ax1.set_xlabel('Machine ID')
ax1.set_ylabel('Downtime Minutes')
ax1.set_title('Pareto Analysis of Machine Downtime')

# Line chart for cumulative percentage
ax2 = ax1.twinx()

ax2.plot(cumulative_percentage.values, color='red', marker='o')

ax2.set_ylabel('Cumulative Percentage')

plt.show()
```
